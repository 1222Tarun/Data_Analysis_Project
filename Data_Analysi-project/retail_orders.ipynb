```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.engine import URL

# Read the CSV file
df = pd.read_csv('orders.csv')

# Rename columns to match MySQL table schema
df.columns = ['Order_Id', 'Order_Date', 'Ship_Mode', 'Segment', 'Country', 'City', 'State', 'Postal_Code', 'Region', 'Category', 'Sub_Category', 'Product_Id', 'cost_price', 'List_Price', 'Quantity', 'Discount_Percent']

# Create the connection URL
connection_url = URL.create(
    drivername='mysql+mysqlconnector',
    username='root',
    password='root',
    host='localhost',
    database='tarun',
    query={'auth_plugin': 'mysql_native_password'}
)

# Create the SQLAlchemy engine for MySQL
engine = create_engine(connection_url)

# Write the DataFrame to the MySQL table
df.to_sql('df_orders', con=engine, index=False, if_exists='append')

print("Data inserted successfully")
```

```python
%%sql
CREATE table orders(   
       [order_id] int primary key
      ,[order_date] date
      ,[ship_mode] varchar(20)
      ,[segment] varchar(20)
      ,[country] varchar(20)
      ,[city] varchar(20)
      ,[state] varchar(20)
      ,[postal_code] varchar(20)
      ,[region] varchar(20)
      ,[category] varchar(20)
      ,[sub_category] varchar(20)
      ,[product_id] varchar(50)
      ,[quantity] int
      ,[discount] decimal(7,2)
      ,[sale_price] decimal(7,2)
      ,[profit] decimal(7,2))
```
